```python
import pandas as pd
```

```python
df = pd.read_csv('./pseudomized_aggregated_ALT.csv')
```

```python
df
```

```python
school_df = df[(df['school'] == '148')]
```

```python
df[(df['subject'] == 'Woordenschat (extra)') & (df['lln.ID.school'] == 3566797)]
```

```python
data = []
for i in school_df['lln.ID.school'].unique():
    classes = school_df[school_df['lln.ID.school'] == i]['classname'].unique()
    data.append({'studentID': i, 'classIDs': classes})
```

```python
student_data = pd.DataFrame(data)
student_data.to_csv('studentData.csv', index=False)
school_df.to_csv('schoolData.csv', index=False)
```
